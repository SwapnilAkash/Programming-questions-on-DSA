```python
class Solution:
    def buddyStrings(self, A: str, B: str) -> bool:
        if not A or not B or (len(A) != len(B)):
            return False
        if len(A) == 2:
            A = A[::-1]
            if A == B:
                return True
            else:
                return False
        c = 0
        if A == B:
            s = set()
            for i in A:
                if i in s:
                    c = 1
                    break
                s.add(i)
            if c == 1:
                return True
            else:
                return False
            
        A = [i for i in A]
        B = [i for i in B]
        j = 0
        
        for i in range(len(A)):
            if A[j] == B[j]:
                del A[j]
                del B[j]
            else:
                j+= 1
            
        if len(A) == 2 and len(A) == len(B):
            A.reverse()
            if A == B:
                return True
            else:
                return False
        else:
            return False
        
```
